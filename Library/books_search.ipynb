## Find books

```python
books = pd.read_pickle("Data/books_with_section.p")
#books["id"] = books["id"].astype(int)
books.dtypes
```

```python
students = pd.read_pickle("Data/students.p")
students
```

```python
def search():
    """
    Start search process and log in
    """
    global studentid
    global filtered_books

    filtered_books = pd.DataFrame(columns= books.columns)

    studentid = int(input("Please enter your student ID."))
    password = int(input("Please enter your passwords."))

    if password == students.loc[students["student_id"] == studentid, "password"][0]:
        filters()
    else: 
        print("Student ID or password incorrect. Please try again.")
        search()
```

```python
def filters():

    global filtered_books
    
    print("Please choose a filter.")
    print("You can choose between genre, year, title, and author or type 'stop' for end filtering.")

    f = input("Please enter the filter of your choice.").lower()

    if f == "genre":
        genre()
        filters()
    elif f == "year":
        year()
        filters()
    elif f ==  "title":
        title()
        filters()
    elif f == "author":
        author()
        filters()
    elif f == "stop":
        stop()
    else:
        print("Invalid entry.")
        filters()

```

```python
def genre(): 
        """
        Filter for specific genre.
        """
        global filtered_books

        print("We have 8 genres:")
        print("""For Arts & Music enter Z, for Business, Economics & Industry enter X, 
            Engineering & Programming enter T, for Kids & Fiction enter Y,
            for Law & Crime enter U, for Medicine enter V,
            for Social Science and Teaching enter W and for Other enter S.""")

        genre = input("Please enter the letter of the genre of your choice.").upper()

        if genre in ["S", "T", "U", "V", "W", "X", "Y", "Z"]:
            g = books[books["section"] == genre]
            filtered_books = pd.concat([filtered_books, g])   
        else:
            print("Invalid entry.")
            genre()
                
    
```

```python
def year():
    """
    Filter for specific publication year.
    """
    global filtered_books

    year = int(input("Please enter the year of publication."))
    
    if year in list(books["year"].unique()):
        y = books[books["year"] == year]
        filtered_books = pd.concat([filtered_books, y])
    
    else:
        print("Invalid entry.")
        year()
```

```python
def title():
    """
    Ask for book title and give back best match and location.
    """
    global filtered_books

    title = input("Please enter a book title or choose for a word.").lower()

    books['title'] = books['title'].str.lower()
    t = books.loc[books['title'].str.contains(title)]
    filtered_books = pd.concat([filtered_books, t])
    
```

```python
def author():
    """
    Filter for author name.
    """
    global filtered_books

    auth = input("Please enter the author's last name.").lower()

    books['authors'] = books['authors'].str.lower()
    a = books.loc[books['authors'].str.contains(auth)]
    filtered_books = pd.concat([filtered_books, a])
    
    
    
```

```python
def stop():
    """
    Stop filtering and display best matching books.
    """
    global ids
    
    ids = list(filtered_books["id"].mode())
    column_names = ["title", "authors", "year", "Genre", "section", "shelf", "id", "description", "lang", "format"]
    ordered = books[column_names]
    
    final = pd.DataFrame(columns = column_names)
    for i in ids:
        xx = ordered.loc[ordered["id"]==i]
        final = pd.concat([final, xx])
    
    display(final)

    found_book()
    
```

```python
def found_book():
    """
    Was the right book found?
    """    
    correct = input("Did you find the correct book?").lower()
    if correct == "yes":
        save_book()
    elif correct == "no":
        print("Please try again.")
        filters()
    else:
        print("Invalid entry. Please type yes or no.")
        found_book()
```

```python
def save_book():
    """
    Save the book that was chosen to student dataframe
    """
    choice = int(input("Please enter the ID of the book of your choice."))
    
    if choice in ids:
        students.loc[students.student_id == studentid, "past_books"] = students.loc[students.student_id == studentid, "past_books"].append(choice)
    else:
        print("Invalid entry.")
        save_book()
```
