## Authors

```python
authors = pd.read_csv("authors.csv")
authors
```

```python
## function to put authors name

def translate_authors(row):
    author_list = []
    stripped = row.strip("[]")
    lis = [int(num) for num in stripped.split(',') if num != '']
    for i in lis: 
        name = authors.loc[authors['author_id'] == i, 'author_name'].values[0]
        author_list.append(name)
    return author_list
```

```python
def translate_formats(row):
    format_list = []
    for i in formats['format_id']:
        if row - i == 0:
            f = formats.loc[formats['format_id'] == i, 'format_name'].values[0]
            format_list.append(f)
    return format_list
```

```python
translate_formats(1.0)
```

```python
books["format"] = books["format"].apply(translate_formats)
```
